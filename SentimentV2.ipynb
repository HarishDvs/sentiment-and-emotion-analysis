```python
import pandas as pd
import numpy as np
import re
import nltk
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.naive_bayes import MultinomialNB
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
train_data.head()
```

```python
test_data.shape
```

```python
train_data.value_counts('Emotion')
```

```python
train_data = train_data.replace({"Emotion":{
    'joy':0,
    'sadness':1,
    'anger': 2,
    'fear': 3,
    'love': 4,
    'surprise': 5}})
test_data = test_data.replace({"Emotion":{
    'joy':0,
    'sadness':1,
    'anger': 2,
    'fear': 3,
    'love': 4,
    'surprise': 5}})
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
train_data['Text'] = train_data['Text'].apply(stemming)
```

```python
train_data.head()
```

```python
text_train = train_data['Text'].values
emotion_train = train_data['Emotion'].values
```

```python
print(emotion_train)
```

```python
test_data['Text'] = test_data['Text'].apply(stemming)
text_test = test_data['Text'].values
emotion_test = test_data['Emotion'].values
```

```python
vectorizer = CountVectorizer(stop_words='english')
text_train = vectorizer.fit_transform(text_train).toarray()
text_test = vectorizer.fit_transform(text_test).toarray()
```

```python
model = MultinomialNB()
model.fit(text_train, emotion_train)
```

```python
train_prediction = model.predict(text_train)
train_acc = accuracy_score(train_prediction, emotion_train)
print(train_acc*100)
```

```python
test_prediction = model.predict(text_test)
test_acc = accuracy_score(train_prediction, emotion_test)
print(test_acc*100)
```
